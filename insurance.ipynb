Sigmoid Function Implementation it returns [0,1]

```python
import math
def sigmoidFunction(x):
    val = 1/(1+math.exp(-x))
    return val
sigmoidFunction(300)
```

tanh function function returns [-1,1]


```python
def tanh(x):
    val = (math.exp(x) - math.exp(-x))/(math.exp(x)+math.exp(-x))
    return val

tanh(-100)
```

relu function implementation it returns max(0,x)

```python
def relu(x):
    return max(0,x)

relu(-1)
```

Implementation of LRelu it returns max(0.1x,x)

```python
def leakyRelu(x):
    return max(0.1*x,x)

leakyRelu(-100)
```

custom gradient descent function

```python
def sigmoidNumpyFunction(x):
    val = 1/(1+np.exp(-x))
    return val
```

```python
def logLossFunction(ytrue, ypredNew):
    epsilon = 1e-15
    ypredNew = np.maximum(ypredNew, epsilon)
    ypredNew = np.minimum(ypredNew, 1 - epsilon)
    loss = -np.mean(ytrue * np.log(ypredNew) + (1 - ytrue) * np.log(1 - ypredNew))
    return loss
```

```python
def gradient_descent(age,y_train,epochs):
    w1 = w2 = 1
    bias = 0
    rate = 0.5
    n = len(age)
    
    for i in range(epochs):
        weighted_sum = w1*age + bias
        y_predicted = sigmoidNumpyFunction(weighted_sum)
        loss = logLossFunction(y_train,y_predicted)
        w1d = (1/n)*np.dot(np.transpose(age),(y_predicted - y_train))
        bias_d = np.mean(y_predicted-y_train)
        w1 = w1 - rate*w1d
        bias = bias - rate*bias_d
        print(f'epoch:{i},w1:{w1},bias:{bias},loss:{loss}')
        
    return w1,bias
        
    
```
